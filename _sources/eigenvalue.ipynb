# **Definisi Eigenvalue, Eigenvector**
**Nilai eigen (eigenvalue)** adalah skalar yang terkait dengan matriks persegi dan vektor eigen, yang menggambarkan bagaimana matriks tersebut merubah vektor tersebut saat dikalikan

**Vektor eigen (eigenvector)** adalah vektor yang, setelah mengalami transformasi linier (biasanya oleh suatu matriks), tetap sejajar dengan vektor asalnya, hanya berubah dalam skala (panjang) saja

$$ A v = λ v $$

```python
import numpy as np
A = np.array([[2,1], [1,2]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print("EigenValues", eigenvalues)
print("EigenVectors", eigenvectors)
```

Diberikan matriks

\[
A = \begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}
\]

Kita ingin mencari nilai eigen (\(\lambda\)) dan vektor eigen (\(v\)) yang memenuhi persamaan

\[
A v = \lambda v
\]

Nilai eigen \(\lambda\) adalah skalar yang membuat persamaan tersebut benar, dan vektor eigen \(v\) adalah vektor tidak nol yang berubah hanya skala ketika dikalikan oleh matriks \(A\).

Untuk mencari nilai eigen, kita selesaikan persamaan karakteristik

\[
\det(A - \lambda I) = 0
\]

dengan \(I\) adalah matriks identitas. Matriks karakteristik adalah

\[
A - \lambda I = \begin{bmatrix}
2 - \lambda & 1 \\
1 & 2 - \lambda
\end{bmatrix}
\]

Menghitung determinan memberikan

\[
(2 - \lambda)(2 - \lambda) - 1 \times 1 = (\lambda^2 - 4\lambda + 3) = 0
\]

Menyelesaikan persamaan kuadrat ini

\[
\lambda^2 - 4\lambda + 3 = 0
\]

dengan rumus kuadrat

\[
\lambda = \frac{4 \pm \sqrt{16 - 12}}{2} = \frac{4 \pm 2}{2}
\]

menghasilkan dua nilai eigen

\[
\lambda_1 = 3, \quad \lambda_2 = 1
\]

Setelah menemukan nilai eigen, kita tentukan vektor eigen dengan menyelesaikan

\[
(A - \lambda I)v = 0
\]

Untuk \(\lambda_1 = 3\), matriksnya menjadi

\[
A - 3I = \begin{bmatrix}
-1 & 1 \\
1 & -1
\end{bmatrix}
\]

Menyelesaikan sistem linear ini menghasilkan vektor eigen

\[
v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
\]

Untuk \(\lambda_2 = 1\), matriksnya menjadi

\[
A - I = \begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}
\]

dan vektor eigen yang diperoleh adalah

\[
v_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}
\]

Untuk membandingkan hasil ini dengan hasil dari perhitungan numerik (misalnya menggunakan NumPy), vektor-vektor eigen biasanya dinormalisasi sehingga memiliki panjang satu, yaitu

\[
v_{\text{normalized}} = \frac{v}{\|v\|}
\]

dengan norma vektor

\[
\|v\| = \sqrt{v_1^2 + v_2^2}
\]

Dalam kasus ini

\[
\|v_1\| = \sqrt{1^2 + 1^2} = \sqrt{2}, \quad
\|v_2\| = \sqrt{1^2 + (-1)^2} = \sqrt{2}
\]

Sehingga vektor eigen yang ternormalisasi adalah

\[
v_1^{norm} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad
v_2^{norm} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ -1 \end{bmatrix}
\]

Dengan demikian, proses ini menjelaskan bagaimana nilai dan vektor eigen dari matriks \(A\) dapat dihitung secara analitik dan dibandingkan dengan hasil numerik dari metode komputer.




---



```python
import numpy as np

# 1. Definisi matriks A
A = np.array([[2, 1],
              [1, 2]])
print("Matriks A:")
print(A)

# 2. Menghitung eigenvalues dan eigenvectors menggunakan numpy
eigenvalues, eigenvectors = np.linalg.eig(A)

# 3. Menampilkan hasil dari numpy
print("\nHasil dari NumPy:")
print("Eigenvalues (nilai eigen):")
print(eigenvalues)
print("Eigenvectors (vektor eigen):")
print(eigenvectors)

# 4. Verifikasi manual - Langkah-langkah mencari eigenvalues
# Matriks karakteristik: |A - λI| = 0
# A - λI = [[2 - λ, 1],
#           [1, 2 - λ]]
# Determinan:
# (2 - λ)(2 - λ) - 1*1 = λ^2 - 4λ + 3 = 0

print("\nLangkah Manual:")
print("Karakteristik persamaan: λ^2 - 4λ + 3 = 0")

# 5. Menyelesaikan persamaan kuadrat λ^2 - 4λ + 3 = 0
# Menggunakan rumus kuadrat:
# λ = [-b ± sqrt(b^2 - 4ac)] / 2a
a = 1
b = -4
c = 3
discriminant = b**2 - 4*a*c
lambda1 = (-b + np.sqrt(discriminant)) / (2*a)
lambda2 = (-b - np.sqrt(discriminant)) / (2*a)

print(f"Eigenvalue 1: {lambda1}")
print(f"Eigenvalue 2: {lambda2}")

# 6. Menentukan eigenvector untuk masing-masing eigenvalue

# Untuk λ = 3:
# (A - 3I)v = 0 → [[-1, 1], [1, -1]]v = 0 → solusi: v1 = [1, 1]
print("\nUntuk λ = 3:")
A_minus_3I = A - 3 * np.eye(2)
print("A - 3I:")
print(A_minus_3I)
print("Solusi vektor eigen: v1 = [1, 1]")

# Untuk λ = 1:
# (A - 1I)v = 0 → [[1, 1], [1, 1]]v = 0 → solusi: v2 = [1, -1]
print("\nUntuk λ = 1:")
A_minus_1I = A - 1 * np.eye(2)
print("A - 1I:")
print(A_minus_1I)
print("Solusi vektor eigen: v2 = [1, -1]")

# 7. Normalisasi vektor eigen untuk dibandingkan dengan numpy
v1 = np.array([1, 1])
v1_normalized = v1 / np.linalg.norm(v1)
v2 = np.array([1, -1])
v2_normalized = v2 / np.linalg.norm(v2)

print("\nNormalisasi manual eigenvectors:")
print("v1 normalized:", v1_normalized)
print("v2 normalized:", v2_normalized)

# 8. Bandingkan hasil manual dengan hasil NumPy
print("\nPerbandingan dengan NumPy:")
print("Eigenvector dari NumPy untuk λ = 3 (kolom 0):", eigenvectors[:, 0])
print("Eigenvector dari NumPy untuk λ = 1 (kolom 1):", eigenvectors[:, 1])
```

### Penjelasan Kode Python: Eigenvalue dan Eigenvector Matriks

1. **Definisi Matriks A**

Kode mendefinisikan matriks:
$$
A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
$$

2. **Menghitung Eigenvalue dan Eigenvector Menggunakan NumPy**

Fungsi `np.linalg.eig(A)` digunakan untuk menghitung eigenvalue ($\lambda$) dan eigenvector ($v$) dari matriks $A$. Hasilnya disimpan dalam variabel:
$$
\text{eigenvalues}, \quad \text{eigenvectors}
$$

3. **Verifikasi Manual: Menentukan Eigenvalue**

Langkah manual:
- Bentuk matriks karakteristik:
$$
A - \lambda I = \begin{bmatrix} 2 - \lambda & 1 \\ 1 & 2 - \lambda \end{bmatrix}
$$
- Determinan:
$$
|A - \lambda I| = (2 - \lambda)(2 - \lambda) - (1 \cdot 1) = \lambda^2 - 4\lambda + 3 = 0
$$
- Selesaikan persamaan kuadrat:
$$
\lambda = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$
Dengan:
$$
a = 1, \quad b = -4, \quad c = 3
$$
Diskriminan:
$$
\Delta = b^2 - 4ac = 16 - 12 = 4
$$
Hasil:
$$
\lambda_1 = 3, \quad \lambda_2 = 1
$$

4. **Menentukan Eigenvector**

- Untuk $\lambda = 3$:
$$
A - 3I = \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix}
$$
Penyelesaian:
$$
v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

- Untuk $\lambda = 1$:
$$
A - I = \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix}
$$
Penyelesaian:
$$
v_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}
$$

5. **Normalisasi Eigenvector**

Untuk membandingkan dengan hasil dari NumPy, eigenvector dinormalisasi:
$$
v_{\text{normalized}} = \frac{v}{\|v\|}
$$
Untuk:
$$
v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad \|v_1\| = \sqrt{1^2 + 1^2} = \sqrt{2}
$$
$$
v_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}, \quad \|v_2\| = \sqrt{1^2 + (-1)^2} = \sqrt{2}
$$

6. **Perbandingan dengan Hasil NumPy**

Hasil manual yang dinormalisasi dibandingkan dengan hasil NumPy:
- Kolom ke-0 dari `eigenvectors` sesuai untuk $\lambda = 3$
- Kolom ke-1 dari `eigenvectors` sesuai untuk $\lambda = 1$


```python
import numpy as np
import matplotlib.pyplot as plt

# Vektor hasil manual
v1 = np.array([2, 1])
v2 = np.array([1, 2])

# Fungsi untuk normalisasi vektor
def normalize(v):
    return v / np.linalg.norm(v)

# Vektor yang sudah dinormalisasi
v1_norm = normalize(v1)
v2_norm = normalize(v2)

print("Vektor [2, 1] setelah dinormalisasi:", v1_norm)
print("Vektor [1, 2] setelah dinormalisasi:", v2_norm)

plt.figure(figsize=(8, 8))
ax = plt.gca()

# Plot vektor asli
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1 = [2,1]')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='v2 = [1,2]')

# Plot vektor yang sudah dinormalisasi (tanpa linestyle karena quiver tidak support linestyle)
ax.quiver(0, 0, v1_norm[0], v1_norm[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1 norm')
ax.quiver(0, 0, v2_norm[0], v2_norm[1], angles='xy', scale_units='xy', scale=1, color='orange', label='v2 norm')

# Atur batas sumbu
plt.xlim(-0.5, 2.5)
plt.ylim(-0.5, 2.5)

plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Diagram Vektor dan Normalisasi Vektor')
plt.legend()
plt.show()
```

# **Definisi Ortonormal dan Ortogonal**

**Pengertian Ortogonal**

Ortogonal adalah istilah dalam matematika yang digunakan untuk menyebut dua buah objek, biasanya vektor, yang saling tegak lurus satu sama lain. Jika kita bayangkan dua garis di bidang datar atau ruang, kedua garis itu dikatakan ortogonal jika sudut di antara mereka tepat 90 derajat, atau dengan kata lain mereka membentuk sudut siku-siku. Dalam konteks aljabar vektor, ketika dua vektor disebut ortogonal, itu berarti tidak ada komponen satu vektor yang “mengarah” ke arah vektor lainnya — mereka berdiri sendiri-sendiri dan tidak saling memengaruhi dalam arah.

Ortogonalitas menjadi konsep yang sangat penting dalam banyak bidang, mulai dari geometri, aljabar linear, analisis data, hingga ilmu komputer, karena ia membantu memisahkan atau memecah suatu masalah menjadi bagian-bagian yang independen atau tidak saling terkait. Misalnya, dalam pemrosesan sinyal atau gambar, basis ortogonal sering digunakan untuk memecah data menjadi komponen-komponen yang mudah diolah. Dalam ruang vektor, konsep ortogonalitas membantu memudahkan perhitungan karena ketika dua vektor ortogonal, hubungan matematis di antara mereka menjadi sederhana.

Selain itu, dua fungsi atau dua vektor ortogonal sering dianggap “tidak berhubungan” atau “independen” satu sama lain. Artinya, satu tidak membawa informasi tentang yang lain. Ini sebabnya ortogonalitas sering digunakan dalam teknik pemisahan sinyal atau dalam algoritma kompresi data, karena kita bisa memecah informasi kompleks menjadi blok-blok sederhana yang tidak saling campur.


---



**Pengertian Ortonormal**

Ortonormal adalah istilah yang menggabungkan dua sifat penting dari sekelompok vektor, yaitu ortogonalitas dan normalitas. Kita sudah tahu bahwa ortogonal berarti saling tegak lurus. Sementara itu, normal berarti setiap vektor memiliki panjang satu, atau sering disebut sebagai vektor satuan. Jadi, sekumpulan vektor dikatakan ortonormal jika setiap pasangan vektor di dalamnya saling ortogonal (tegak lurus) dan setiap vektor di dalam kelompok itu memiliki panjang atau norma satu.

Mengapa ortonormal penting? Karena sekumpulan vektor ortonormal membentuk apa yang disebut sebagai basis ortonormal di dalam suatu ruang vektor. Basis ortonormal mempermudah hampir semua perhitungan dalam ruang tersebut. Misalnya, untuk mencari proyeksi suatu vektor ke dalam basis ortonormal, kita tidak perlu melakukan perhitungan yang rumit. Selain itu, transformasi yang dinyatakan menggunakan basis ortonormal biasanya menjaga jarak dan sudut, sehingga sifat-sifat geometri tetap terpelihara. Ini sangat berguna dalam bidang seperti komputer grafis, machine learning, statistika, fisika, dan banyak lagi.

Kita bisa membayangkan ortonormal sebagai tim yang sangat teratur dan efisien. Setiap anggota tim (vektor) punya arah yang berbeda dengan anggota lainnya (tegak lurus) dan semuanya bekerja dengan kekuatan penuh yang sama (panjang satu). Tidak ada yang saling tumpang tindih atau saling mengurangi satu sama lain. Karena sifat ini, ortonormal sering digunakan dalam perhitungan numerik untuk menjaga kestabilan algoritma, seperti dalam dekomposisi matriks dan analisis data dimensi tinggi.

```python
import numpy as np

# Definisi matriks A (simetris)
A = np.array([[2, 1],
              [1, 2]])

# Hitung eigenvalues dan eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors (belum normalisasi):")
print(eigenvectors)

# Normalisasi setiap kolom vektor eigen (jadi panjangnya 1)
eigenvectors_orthonormal = eigenvectors / np.linalg.norm(eigenvectors, axis=0)

print("\nEigenvectors setelah normalisasi (orthonormal):")
print(eigenvectors_orthonormal)

# Cek orthonormalitas:
print("\nCek dot product antar vektor eigen (harus dekat 0):")
dot_product = np.dot(eigenvectors_orthonormal[:, 0], eigenvectors_orthonormal[:, 1])
print(f"Dot product vektor 1 dan 2: {dot_product:.6f}")

print("\nCek norma (panjang) tiap vektor eigen (harus 1):")
norm1 = np.linalg.norm(eigenvectors_orthonormal[:, 0])
norm2 = np.linalg.norm(eigenvectors_orthonormal[:, 1])
print(f"Norm vektor 1: {norm1:.6f}")
print(f"Norm vektor 2: {norm2:.6f}")
```

### Penjelasan Lengkap Kode Python: Eigenvalue dan Eigenvector Matriks

---

#### 1. **Definisi Matriks \( A \)**

Matriks yang akan dianalisis adalah:

\[
A = \begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}
\]

Matriks ini adalah matriks simetris, yang berarti \( A = A^T \).

---

#### 2. **Menghitung Eigenvalues dan Eigenvectors dengan NumPy**

Kode Python menggunakan fungsi \texttt{np.linalg.eig(A)} untuk menghitung:

- **Eigenvalues** (\(\lambda\)) dari matriks \(A\), yaitu skalar yang memenuhi persamaan  
  \[
  A v = \lambda v
  \]
- **Eigenvectors** (\(v\)), yaitu vektor bukan nol yang memenuhi persamaan tersebut.

Fungsi tersebut mengembalikan:

\[
\text{eigenvalues}, \quad \text{eigenvectors}
\]

dengan eigenvectors disusun sebagai kolom-kolom pada matriks hasil.

---

#### 3. **Verifikasi Manual: Menentukan Eigenvalues**

Untuk memahami prosesnya secara manual, kita dapat:

- Bentuk matriks karakteristik \( A - \lambda I \), di mana \(I\) adalah matriks identitas:

\[
A - \lambda I = \begin{bmatrix}
2 - \lambda & 1 \\
1 & 2 - \lambda
\end{bmatrix}
\]

- Hitung determinan matriks karakteristik dan setarakan dengan nol:

\[
|A - \lambda I| = (2 - \lambda)(2 - \lambda) - (1 \times 1) = \lambda^2 - 4\lambda + 3 = 0
\]

- Selesaikan persamaan kuadrat:

\[
\lambda = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
\]

dengan

\[
a = 1, \quad b = -4, \quad c = 3
\]

- Hitung diskriminan:

\[
\Delta = b^2 - 4ac = 16 - 12 = 4
\]

- Hasil solusi:

\[
\lambda_1 = 3, \quad \lambda_2 = 1
\]

---

#### 4. **Menentukan Eigenvectors untuk Setiap Eigenvalue**

- Untuk \(\lambda_1 = 3\), substitusi ke matriks \( A - 3I \):

\[
A - 3I = \begin{bmatrix}
2-3 & 1 \\
1 & 2-3
\end{bmatrix} =
\begin{bmatrix}
-1 & 1 \\
1 & -1
\end{bmatrix}
\]

- Cari vektor eigen \( v_1 \) yang memenuhi:

\[
(A - 3I)v_1 = 0
\]

Penyelesaian menunjukkan:

\[
v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
\]

- Untuk \(\lambda_2 = 1\), substitusi ke matriks \( A - I \):

\[
A - I = \begin{bmatrix}
2-1 & 1 \\
1 & 2-1
\end{bmatrix} =
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}
\]

- Cari vektor eigen \( v_2 \) yang memenuhi:

\[
(A - I)v_2 = 0
\]

Penyelesaian menunjukkan:

\[
v_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}
\]

---

#### 5. **Normalisasi Eigenvectors**

Untuk membandingkan hasil dengan output NumPy, eigenvectors perlu dinormalisasi sehingga panjangnya menjadi 1. Normalisasi dilakukan dengan rumus:

\[
v_{\text{normalized}} = \frac{v}{\|v\|}
\]

dengan norma \(\|v\|\) dihitung sebagai:

\[
\|v\| = \sqrt{v_1^2 + v_2^2}
\]

Khusus untuk:

\[
v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad
\|v_1\| = \sqrt{1^2 + 1^2} = \sqrt{2}
\]

\[
v_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}, \quad
\|v_2\| = \sqrt{1^2 + (-1)^2} = \sqrt{2}
\]

Sehingga eigenvectors yang dinormalisasi menjadi:

\[
v_1^{norm} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \quad
v_2^{norm} = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}
\]

---

#### 6. **Perbandingan dengan Hasil NumPy**

- Kolom pertama dari matriks \texttt{eigenvectors} NumPy sesuai dengan eigenvector untuk \(\lambda_1 = 3\).
- Kolom kedua dari matriks \texttt{eigenvectors} NumPy sesuai dengan eigenvector untuk \(\lambda_2 = 1\).

---

#### 7. **Kesimpulan**

Kode Python menggunakan \texttt{np.linalg.eig} secara efisien mengembalikan eigenvalues dan eigenvectors. Eigenvectors yang dihasilkan biasanya sudah dalam bentuk ter-normalisasi, sehingga siap dipakai untuk analisis lebih lanjut seperti dekomposisi spektral, transformasi basis, atau aplikasi lain dalam ilmu data dan fisika.
